## Model Fitting

```python
import numpy as np

# Base Regularization class 
class Regularization:
    def __init__(self,penalty = None, lambda_ = 0.0 ):
        self.penalty = penalty
        self.lambda_ = lambda_

    def soft_thresholding(self, SXY, SXX):
        Threshold = self.lambda_ / 2
        if SXY > Threshold:
            return (SXY - Threshold) / SXX
        elif SXY < -Threshold:
            return (SXY + Threshold) / SXX
        else:
            return 0    
        
    def gradient_Regularize(self, weights):
            if self.penalty == 'l1':
                return self.lambda_ * np.sign(weights)
            elif self.penalty == 'l2':
                return 2* self.lambda_* weights
            else:
                return 0.0
```

## Numerical View Implementation

```python
# Inherit from Regularization class to implement Linear Regression with regularization

class LinearRegressionNumericalView(Regularization):
    def __init__ (self, penalty=None, lambda_=0.0):
        super().__init__(penalty, lambda_)
        self.a0 = 0 
        self.a1 = 0
        self.SSE = None
        self.MSE = None

    @staticmethod
    def sse(y,y_hat):
        return np.sum((y - y_hat) **2)
    
    def predict(self , X):  
        # Y = a0 + a1*X >> Linear regression equation
        X = np.asarray(X , dtype=float).reshape(-1)
        return self.a0 + self.a1 * X
    
    def fit(self, X, Y):
        X= np.asarray(X , dtype=float).reshape(-1)
        Y= np.asarray(Y , dtype=float).reshape(-1)
        n=len(X)
        
        # Cala Summations 
        SX= np.sum(X)
        SY= np.sum(Y)
        SXX= np.sum(X**2)
        SXY= np.sum(X*Y)

        if SXX == 0:
            raise ValueError("Variance of X is zero. Cannot fit a linear model.")

        # Calculate the penalty term based on the specified regularization type
        if self.penalty == 'l2' :

            penalty_value =  n * self.lambda_  # Ridge regression penalty term (λ * n)
            A = np.array ( [[n , SX] , 
                        [SX , SXX + penalty_value]] , dtype=float)
            b = np.array ([SY , SXY], dtype=float )
            self.a0 , self.a1 = np.linalg.solve(A, b)

        elif self.penalty == 'l1' :
            self.a1 = self.soft_thresholding(SXY, SXX)

            # Intercept (a0) calculation
            X_bar = np.mean(X)
            Y_bar = np.mean(Y)
            self.a0 = Y_bar - self.a1 * X_bar

        else:

            # Simple Linear Regression (No Regularization)
            A = np.array ( [[n , SX] ,
                    [SX , SXX]] , dtype=float)
            
            b = np.array ([SY , SXY], dtype=float )
            
            # Solve 
            self.a0 , self.a1 = np.linalg.solve(A,b)

        # Final Model 
        y_hat = self.predict(X)
        self.SSE = self.sse( Y, y_hat)
        self.MSE = self.SSE / n

        # print("Final Model:")
        # print(f"Penalty: {self.penalty}, λ = {self.lambda_}")
        # print(f"ŷ = {self.a0:.4f} + {self.a1:.4f} x\n")
        # print(f"SSE = {self.SSE}")
        # print(f"MSE = {self.MSE}")
        # print(f"RMSE = {np.sqrt(self.MSE):.4f}")

        return self
    
# Example usage
X_train = train_cleaned['Adult Mortality']
Y_train = train_cleaned['Life expectancy']

Rmodel = LinearRegressionNumericalView('l2', 0.001)
Rmodel.fit(X_train, Y_train)
print('-'*50)
print(f'Coefficients of Ridge Model: a0 = {Rmodel.a0:.4f}, a1 = {Rmodel.a1:.4f}')
print(f'SSE of Ridge Model: {Rmodel.SSE:.4f}')
print(f'MSE of Ridge Model: {Rmodel.MSE:.4f}')
print(f'RMSE of Ridge Model: {np.sqrt(Rmodel.MSE):.4f}')

print('*'*80)

Lmodel = LinearRegressionNumericalView('l1', 0.001)
Lmodel.fit(X_train, Y_train)
print('-'*50)
print(f'Coefficients of Lasso Model: a0 = {Lmodel.a0:.4f}, a1 = {Lmodel.a1:.4f}')
print(f'SSE of Lasso Model: {Lmodel.SSE:.4f}')
print(f'MSE of Lasso Model: {Lmodel.MSE:.4f}')
print(f'RMSE of Lasso Model: {np.sqrt(Lmodel.MSE):.4f}')

print('*'*80)

model = LinearRegressionNumericalView()
model.fit(X_train, Y_train)
print('-'*50)
print(f'Coefficients of Simple Model: a0 = {model.a0:.4f}, a1 = {model.a1:.4f}')
print(f'SSE of Simple Model: {model.SSE:.4f}')
print(f'MSE of Simple Model: {model.MSE:.4f}')
print(f'RMSE of Simple Model: {np.sqrt(model.MSE):.4f}')
```

## Statistical View Implementation

```python
class LinearRegressionstatistical(Regularization):
    def __init__(self, penalty=None, lambda_=0.0):
        super().__init__(penalty, lambda_)
        self.B0 = None
        self.B1 = None
        self.MSE = None

    def fit(self, X, Y):
        X = np.array(X).reshape(-1) 
        Y = np.array(Y).reshape(-1)
        n = len(X)

        X_bar = X.mean() 
        Y_bar = Y.mean() 

        # Calculate SXX and SXY
        SXX = np.sum(X ** 2) - (n * (X_bar ** 2))
        SXY = np.sum(X * Y) - (n * X_bar * Y_bar)

        if SXX == 0 :
            raise ValueError("Cannot divide by zero, variance of X is zero")
        
        # Regularization 
        if self.penalty == 'l2' :
            penalty_value = self.lambda_ * n
            self.B1 = SXY / (SXX + penalty_value)  # Ridge penalty added to the slope
            
        elif self.penalty == 'l1' :
            # Lasso regression with soft thresholding
            self.B1 = self.soft_thresholding(SXY, SXX)
        else:
            # Simple linear regression without regularization
            self.B1 = SXY / SXX
            
        # Calculate intercept
        self.B0 = Y_bar - self.B1 * X_bar

        # Calculate estimated error
        Y_hat = self.predict(X)
        e = Y - Y_hat
        SSE = np.sum((e) ** 2)
        self.MSE = SSE / n 

    def predict(self , X) :
        if self.B0 is None or self.B1 is None:
            raise ValueError ("Model has not been fitted Yet")
        
        X = np.array(X).reshape(-1)
        return self.B0 + self.B1 * X 
    
# Example usage
X_train = train_cleaned['Adult Mortality']
Y_train = train_cleaned['Life expectancy']

Rmodel_stat = LinearRegressionstatistical('l2', 0.001)
Rmodel_stat.fit(X_train, Y_train)
print("Evaluation of the Ridge model")
print("------------------------")
print(f'Line of best fit is: y = {Rmodel_stat.B0:.3f} + {Rmodel_stat.B1:.3f} x')
print(f'Mean squared error is: {Rmodel_stat.MSE:.3f}')
print(f'RMSE = {np.sqrt(Rmodel_stat.MSE):.3f}')

print('*'*80)

Lmodel_stat = LinearRegressionstatistical('l1', 0.001)
Lmodel_stat.fit(X_train, Y_train)
print("Evaluation of the Lasso model")
print("------------------------")
print(f'Line of best fit is: y = {Lmodel_stat.B0:.3f} + {Lmodel_stat.B1:.3f} x')
print(f'Mean squared error is: {Lmodel_stat.MSE:.3f}')
print(f'RMSE = {np.sqrt(Lmodel_stat.MSE):.3f}')

print('*'*80)

model_stat = LinearRegressionstatistical()
model_stat.fit(X_train, Y_train)
print("Evaluation of the Simple model")
print("------------------------")
print(f'Line of best fit is: y = {model_stat.B0:.3f} + {model_stat.B1:.3f} x')
print(f'Mean squared error is: {model_stat.MSE:.3f}')
print(f'RMSE = {np.sqrt(model_stat.MSE):.3f}')
```

### Why do we use Soft Thresholding in Lasso Regression?

In Lasso Regression, we add a penalty that pushes some weights toward zero.  
Unlike other regression methods, Lasso can completely remove some features by setting their weights to zero.

Because of this, we use a simple rule called **Soft Thresholding** to update the weights.

### Simple idea:

- If a weight is small → it becomes **0**
- If a weight is large → it is **reduced**

### Why is this useful?

- It helps remove useless features
- It makes the model simpler
- It reduces overfitting (helps the model generalize better)

### In short:

Soft Thresholding helps Lasso automatically choose the most important features and ignore the rest.

## Machine Learning Implementation With Gradient Descent

```python
class MultipleLinearRegression(Regularization):
    def __init__(self , penalty=None, lambda_=0.0 , n_features=1):
        super().__init__(penalty, lambda_)
        self.W0 = None # Intercept
        self.W = np.zeros(n_features) # weights for each feature
        self.SSE = float('inf')
        self.MSE = None
        self.loss_history = [] # to track the loss during training

    def fit(self , X ,Y , Learning_Rate = 0.01 , epoches = 1000 , tolerance = 1e-6):
        # initialate Parameters 
        self.W0 = 0
        self.X = np.array(X)
        self.Y = np.array(Y).reshape(-1)
        self.n_samples , self.n_features = self.X.shape
        self.W = np.zeros(self.n_features) # weights for each feature

        # Gradient Descent
        for _ in range(epoches):
            y_hat = self.W0 + np.dot(self.X, self.W) # predicted values
            e = self.Y - y_hat

            # Gradients
            # J = (1/2n) * sum((y - y_hat)^2) 
            dW0 = (-1/self.n_samples) * (np.sum(e)) 
            dW = (-1/self.n_samples) * (np.dot(self.X.T, e))
			
            # Regularization Gradient
            if self.penalty in ['l1', 'l2']:
                dW += self.gradient_Regularize(self.W) 
            else:
                dW += 0.0

            # Update 
            self.W0 -= Learning_Rate * dW0
            self.W -= Learning_Rate * dW

            # Compute new SSE 
            y_hat = self.W0 + np.dot(self.X, self.W) # predicted values
            new_SSE = np.sum( (self.Y - y_hat) **2)
            self.loss_history.append(new_SSE)

            # Convergence Check 
            if abs(new_SSE - self.SSE) < tolerance:
                self.SSE = new_SSE
                break
            self.SSE = new_SSE

        # Calc MSE 
        self.MSE = self.SSE / self.n_samples 

    def predict(self ,X) :
        if self.W0 is None or self.W is None :
            raise ValueError("The Model has been fitted yet")  
        X = np.array(X)
        return self.W0 + np.dot(X, self.W) # The value of y_hat (X is the input)
    

#Example usage
X_train_multi = train_cleaned[['Adult Mortality','infant deaths','Measles','HIV/AIDS']]
Y_train_multi = train_cleaned['Life expectancy']

multi_Rmodel = MultipleLinearRegression(penalty='l2', lambda_=0.01, n_features=X_train_multi.shape[1])
multi_Rmodel.fit(X_train_multi, Y_train_multi, 0.01 )
print("Evaluation of the Regularized Ridge Multiple Linear Regression model")
print("------------------------")
print(f'Coefficients: W0 (intercept) = {multi_Rmodel.W0:.3f}, W (weights) = {multi_Rmodel.W}')
print(f'Mean squared error: {multi_Rmodel.MSE:.3f}')

print('*'*80)

multi_Lmodel = MultipleLinearRegression(penalty='l1', lambda_=0.01, n_features=X_train_multi.shape[1])
try:
    multi_Lmodel.fit(X_train_multi, Y_train_multi, 0.01)
    print("Evaluation of the Regularized Lasso Multiple Linear Regression model")
    print("------------------------")
    print(f'Coefficients: W0 (intercept) = {multi_Lmodel.W0:.3f}, W (weights) = {multi_Lmodel.W}')
    print(f'Mean squared error: {multi_Lmodel.MSE:.3f}')
except ValueError as e:
    print(f"Error during Lasso regression: {e}")


print('*'*80)

model_multi = MultipleLinearRegression(n_features=X_train_multi.shape[1])
model_multi.fit(X_train_multi, Y_train_multi, 0.01)
print("Evaluation of the Multiple Linear Regression model")
print("------------------------")
print(f'Coefficients: W0 (intercept) = {model_multi.W0:.3f}, W (weights) = {model_multi.W}')
print(f'Mean squared error: {model_multi.MSE:.3f}')

print('*'*80)

X_single = train_cleaned[['Adult Mortality']]
model_simple = MultipleLinearRegression()
model_simple.fit(X_single, Y_train_multi, 0.01)
print("Evaluation of the Simple Linear Regression model")
print("------------------------")
print(f'Coefficients: W0 (intercept) = {model_simple.W0:.3f}, W (weights) = {model_simple.W}')
print(f'Mean squared error: {model_simple.MSE:.3f}')
```

# Model Comparison Section

In this section we will compute R-squared and Adjusted R-squared from scratch using only NumPy, then build a full comparison table across all model specifications and regularization methods, rank the models by performance, and provide a written interpretation of the results.

## R-squared from Scratch

We implement R-squared manually using NumPy without using any machine learning library.

The formula is:

    R2 = 1 - (SS_res / SS_tot)

Where:
- SS_res = sum of squared residuals = sum((y_true - y_pred)^2)
- SS_tot = total sum of squares     = sum((y_true - mean(y_true))^2)

R-squared tells us what proportion of the variance in the target variable is explained by the model. A value close to 1 means the model explains the data well.

```python
import numpy as np
import pandas as pd

# R-squared measures how well the model explains the variance in the target variable
# Formula: R2 = 1 - (SS_res / SS_tot)
# SS_res = sum of squared differences between actual and predicted values
# SS_tot = sum of squared differences between actual values and their mean
def r_squared(y_true, y_pred):
    y_true  = np.array(y_true)
    y_pred  = np.array(y_pred)
    y_mean  = np.mean(y_true)                    # mean of actual values
    SS_res  = np.sum((y_true - y_pred) ** 2)     # residual sum of squares
    SS_tot  = np.sum((y_true - y_mean) ** 2)     # total sum of squares
    return 1 - (SS_res / SS_tot)
```

## Adjusted R-squared from Scratch

Adjusted R-squared is a modified version of R-squared that penalizes adding predictors that do not actually improve the model.

The formula is:

    Adjusted R2 = 1 - [(1 - R2) * (n - 1) / (n - p - 1)]

Where:
- n = number of samples
- p = number of predictors used in the model

Unlike R-squared, Adjusted R-squared can decrease if we add a predictor that does not help, which makes it a more reliable metric for comparing models with different numbers of features.

```python
# Adjusted R-squared penalizes the model for adding predictors that do not improve performance
# Formula: Adjusted R2 = 1 - [(1 - R2) * (n - 1) / (n - p - 1)]
# n = number of samples, p = number of predictors
def adjusted_r_squared(y_true, y_pred, p):
    y_true = np.array(y_true)
    n      = len(y_true)                       # number of observations
    r2     = r_squared(y_true, y_pred)         # call the r_squared function we defined above
    adj_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
    return adj_r2
```

## Defining Model Configurations

Here we define the three model specifications and the two regularization methods that will be used to build the comparison table.

The three model specifications are:
- Best Single: uses only the single best predictor (HIV/AIDS)
- Full Model: uses all available features after preprocessing
- Domain Model: uses 4 features selected based on domain knowledge

Each specification will be tested with both Ridge (L2) and Lasso (L1) regularization.

```python
# Each tuple contains: (model name, train features, test features, number of predictors)
# These were defined earlier in the notebook by the team
model_configs = [
    ("Best Single",  X_train_best,   X_test_best,   1),
    ("Full Model",   X_train_full,   X_test_full,   X_train.shape[1]),
    ("Domain Model", X_train_domain, X_test_domain, len(domain_features)),
]

# The two regularization methods we will compare
reg_configs = [
    ("Ridge", "l2"),
    ("Lasso", "l1"),
]
```

## Building the Comparison Table

We loop over all model specifications and regularization methods, fit each model, generate predictions on the test set, then compute MSE, R-squared, and Adjusted R-squared for each combination.

The result is a pandas DataFrame where:
- Rows represent each model specification and regularization combination
- Columns are the evaluation metrics: MSE, R-squared, Adjusted R-squared, and number of predictors

```python
comparison_rows = []

# Loop over every combination of model spec and regularization method
for model_name, X_tr, X_te, p in model_configs:
    for reg_name, penalty in reg_configs:

        # Fit the model using the MultipleLinearRegression class defined earlier
        mdl = MultipleLinearRegression(
            penalty=penalty,
            lambda_=0.01,
            n_features=X_tr.shape[1]
        )
        mdl.fit(X_tr, y_train, 0.01, epoches=2000)

        # Get predictions on the test set
        y_pred = mdl.predict(X_te)
        y_true = np.array(y_test)

        # Compute all metrics from scratch
        mse    = np.mean((y_true - y_pred) ** 2)
        r2     = r_squared(y_true, y_pred)
        adj_r2 = adjusted_r_squared(y_true, y_pred, p)

        # Append the results as a row in the table
        comparison_rows.append({
            "Model Spec"     : model_name,
            "Regularization" : reg_name,
            "# Predictors"   : p,
            "MSE"            : round(mse,    4),
            "R2"             : round(r2,     4),
            "Adjusted R2"    : round(adj_r2, 4),
        })

# Build the final DataFrame
comparison_df = pd.DataFrame(comparison_rows)
print(comparison_df.to_string(index=False))
```

## Ranking Models by Performance

We rank the models by Adjusted R-squared in descending order because Adjusted R-squared is the most reliable metric for comparing models with different numbers of predictors. It rewards the model for explaining variance while penalizing unnecessary complexity.

```python
# Sort by Adjusted R2 descending so the best model appears at the top
comparison_df_ranked = comparison_df.sort_values(
    by="Adjusted R2", ascending=False
).reset_index(drop=True)

# Add a rank column starting from 1
comparison_df_ranked.insert(0, "Rank", range(1, len(comparison_df_ranked) + 1))

print("Model Comparison Table - Ranked by Adjusted R2")
print("=" * 70)
print(comparison_df_ranked.to_string(index=False))
print("=" * 70)
```

|---|---|---|

| R2 | 1 - SS_res/SS_tot | Proportion of variance explained, higher is better |

### Results Discussion

**Best Model: Full Model with Ridge regularization**

The Full Model using Ridge regularization achieved the lowest MSE and the highest Adjusted R-squared across all six configurations. This makes sense because the dataset contains 14 informative features after preprocessing, and Ridge keeps all of them by shrinking coefficients towards zero rather than eliminating them entirely. Since most features in this dataset have a genuine relationship with life expectancy, keeping all of them leads to better predictions.

**Full Model with Lasso**

The Full Model with Lasso came in second place with a very similar performance to Ridge. The small gap exists because Lasso drives some coefficients to exactly zero, effectively removing features. On this dataset, most features are useful, so removing any of them slightly hurts performance.

**Domain Model (Ridge and Lasso)**

The Domain Model uses only 4 features selected based on domain knowledge: Adult Mortality, HIV/AIDS, Income composition of resources, and Schooling. It performed moderately well and is only slightly worse than the Full Model. This makes it a strong candidate when interpretability is important, for example in a policy or healthcare report where stakeholders need to understand the model.

Using only HIV/AIDS as a single predictor gave the worst performance by a large margin. Even though HIV/AIDS has the highest individual correlation with life expectancy, one predictor alone cannot capture the full complexity of the target variable. Both Ridge and Lasso performed identically here since there is only one coefficient to regularize.

| Regularization | Behavior | Result on this Dataset |
|---|---|---|
| Ridge (L2) | Shrinks all coefficients, keeps all features | Slightly better because all features are informative |
| Lasso (L1) | Can zero out coefficients, removes some features | Slightly worse because removing features hurts here |

## Cross validation 

```python
from sklearn.model_selection import KFold

def cross_validation(X, y, model_class, lambdas, penalty):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    
    best_lambda = None
    best_mse = float('inf')
    best_metrics = {}

    for lambda_ in lambdas:
        mse_list = []
        R2_list = []

        for train_idx, test_idx in kf.split(X):
            X_train_cv, X_test_cv = X.iloc[train_idx], X.iloc[test_idx]
            y_train_cv, y_test_cv = y.iloc[train_idx], y.iloc[test_idx]

            # Create model
            if model_class == MultipleLinearRegression:
                model = model_class(penalty=penalty, lambda_=lambda_, n_features=X.shape[1])
                model.fit(X_train_cv, y_train_cv, Learning_Rate=0.01, epoches=2000)

            else:
                model = model_class(penalty=penalty, lambda_=lambda_)
                model.fit(X_train_cv, y_train_cv)

            y_pred = model.predict(X_test_cv)

            # Metrics
            mse = np.mean((y_test_cv - y_pred) ** 2)
            mse_list.append(mse)

            SST = np.sum((y_test_cv - np.mean(y_test_cv))**2)
            SSR = np.sum((y_test_cv - y_pred)**2)
            R2 = 1 - (SSR / SST)
            R2_list.append(R2)
        

        avg_mse = np.mean(mse_list)
        avg_r2 = np.mean(R2_list)
        rmse = np.sqrt(avg_mse)
        
        # Choose the best lambda based on MSE
        if avg_mse < best_mse:
            best_mse = avg_mse
            best_lambda = lambda_
            best_metrics = {
                "MSE": avg_mse,
                "RMSE": rmse,
                "R2": avg_r2,
            }        

    return best_lambda, best_metrics



# Define lambda values to test
lambda_values = [0.001, 0.01, 0.1, 1]

# [1] Statistical
best_lambda_stat, metrics_stat = cross_validation(
    X_train_best, y_train, LinearRegressionstatistical, lambda_values, 'l2'
)

# [2] Numerical
best_lambda_num, metrics_num = cross_validation(
    X_train_best, y_train, LinearRegressionNumericalView, lambda_values, 'l2'
)


# [3] Multiple
best_lambda_multi, metrics_multi = cross_validation(
    X_train, y_train, MultipleLinearRegression, lambda_values, 'l2'
)


print("Statistical Model:")
print(best_lambda_stat, metrics_stat)

print("\nNumerical Model:")
print(best_lambda_num, metrics_num)

print("\nMultiple Model:")
print(best_lambda_multi, metrics_multi)
```


In this part, we used **K-Fold Cross Validation** to evaluate our models and choose the best value of λ (lambda).

### What did we do?

- We split the data into **5 folds** using K-Fold.
- In each iteration:
  - We trained the model on 4 folds
  - Tested it on the remaining fold
- This process is repeated 5 times so every part of the data is used for testing once.

### Why did we do this?

- To get a more reliable evaluation (instead of depending on one split)
- To avoid overfitting
- To choose the best λ for regularization

### What did we test?

We tested different values of λ:
- 0.001, 0.01, 0.1, 1

For each value, we calculated:
- **MSE** (Mean Squared Error)
- **RMSE** (Root Mean Squared Error)
- **R² Score**

### How did we choose the best model?

- We selected the λ that gives the **lowest MSE**

### Results:

- Statistical Model → Best λ = 0.001  
- Numerical Model → Best λ = 0.001  
- Multiple Model → Best λ = 0.01  

### Key Insight:

- The **Multiple Linear Regression model performed best**, with:
  - Lower MSE
  - Higher R² score

This shows that using multiple features improves the model performance compared to using a single feature.
